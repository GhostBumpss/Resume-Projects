```python
import pandas as pd
import ast
```

```python
offers_df = pd.read_csv("./offers.csv", header=0)
offers_df.head(n=5)
```

```python
offers_df["channels"] = offers_df["channels"].apply(ast.literal_eval)
adjusted_df = offers_df.explode('channels')
adjusted_df.head(5)
```

```python
adjusted_offers_file_path = "./adjusted_offers.csv"
adjusted_df.to_csv(adjusted_offers_file_path, index=False, header=True)
```
